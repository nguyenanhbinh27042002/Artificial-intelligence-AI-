```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
url = '/content/mtcars.csv'
data = pd.read_csv(url)
data.hist()
plt.show
```

```python
correlation = data.corr() 
print(correlation)
```

mpg số dặm di chuyển trên một gallon
disp thể tích buồng đốt 
hp mã lực 
wt : 

```python
import seaborn as sns 
sns.heatmap(correlation,annot=True)
```

```python
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['wt']),y=data['mpg']) # ham day cho AI hoc dieu chinh lai
print(model.intercept_)
print(model.coef_)
# pt hoi quy : y =coef*x+intercept =>> y =a*x+b

# do chinh xac 
model.score(X=pd.DataFrame(data['wt']),y=data['mpg']) # 75% 
```

```python
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color = 'black',xlim=(0.7))
y_pred = model.predict(X=pd.DataFrame(data['wt']))
# hoi quy ham bac 1
plt.plot(data['wt'],y_pred,color='blue') # duong hoi quy bieu dien moi quan he truc x va y 
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model = linear_model.LinearRegression()
# bac 2 
X = pd.DataFrame([data['wt'],data['wt']**2]).T # T chuyen vi 
model.fit(X,y=data['mpg'])
print('Intercept cua mo hinh :',model.intercept_)
print('Coefficient cua mo hinh :',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,y=data['mpg']))

```

```python
import numpy as np
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color = 'red',xlim=(0.7))
range =np.arange(1.5,5.5,0.1)
predictor = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X=predictor)
plt.plot(range,y_pred,color='blue')
```
