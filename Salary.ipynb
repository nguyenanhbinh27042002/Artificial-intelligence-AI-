Salary and Experience 

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
url = '/content/Salary_Data.csv'
data = pd.read_csv(url)
data.hist()
plt.show
```

```python
correlation = data.corr() 
print(correlation)
```

```python
import seaborn as sns 
sns.pairplot(data)
#sns.heatmap(correlation,annot=True)
```

```python
data.plot(kind='scatter',x='YearsExperience',y='Salary',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['YearsExperience']),y=data['Salary']) # ham day cho AI hoc dieu chinh lai
print(model.intercept_)
print(model.coef_)
# pt hoi quy : y =coef*x+intercept =>> y =a*x+b

# do chinh xac 
model.score(X=pd.DataFrame(data['YearsExperience']),y=data['Salary']) # 95.69  
```

```python
data.plot(kind='scatter',x='YearsExperience',y='Salary',figsize=(9,9),color = 'black',xlim=(0.7))
y_pred = model.predict(X=pd.DataFrame(data['YearsExperience']))
# hoi quy ham bac 1
plt.plot(data['YearsExperience'],y_pred,color='blue') # duong hoi quy bieu dien moi quan he truc x va y 
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
# bac 2
X=pd.DataFrame([data['Salary'],data['Salary']**2]).T
y=data['YearsExperience']
model.fit(X,y)
# bac 2 
X = pd.DataFrame([data['Salary'],data['Salary']**2]).T # T chuyen vi )
print('Intercept cua mo hinh :',model.intercept_)
print('Coefficient cua mo hinh :',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,y=data['YearsExperience']))
```

```python
import numpy as np
data.plot(kind='scatter',x='Salary',y='YearsExperience',figsize=(9,9),color = 'red',xlim=(0.7))
range =np.arange(data['Salary'].min(),data['Salary'].max(),10) # coi lai cho ve do thi 
predictor = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X=predictor)
plt.plot(range,y_pred,color='blue')
```
