Position_Salaries


```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
url = '/content/Position_Salaries.csv'
data = pd.read_csv(url)
data.hist()
plt.show
```

```python
correlation = data.corr() 
print(correlation)
```

```python
import seaborn as sns 
sns.pairplot(data)
#sns.heatmap(correlation,annot=True)
```

```python
data.plot(kind='scatter',x='Salary',y='Level',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
X=pd.DataFrame(data['Salary'])
y=data['Level']
model.fit(X,y) # ham day cho AI hoc dieu chinh lai
print(model.intercept_)
print(model.coef_)

data.plot(kind='scatter',x='Salary',y='Level',figsize=(9,9),color = 'black',xlim=(0.7))
y_pred = model.predict(X=pd.DataFrame(data['Salary']))
# hoi quy ham bac 1
plt.plot(data['Salary'],y_pred,color='blue')
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
# bac 2
X=pd.DataFrame([data['Salary'],data['Salary']**2]).T
y=data['Level']
model.fit(X,y)
# bac 2 
X = pd.DataFrame([data['Salary'],data['Salary']**2]).T # T chuyen vi )
print('Intercept cua mo hinh :',model.intercept_)
print('Coefficient cua mo hinh :',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,y=data['Level']))
```

```python
import numpy as np
data.plot(kind='scatter',x='Salary',y='Level',figsize=(9,9),color = 'red',xlim=(0.7))
range =np.arange(data['Salary'].min(),data['Salary'].max(),10) # coi lai cho ve do thi 
predictor = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X=predictor)
plt.plot(range,y_pred,color='blue')
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
# bac 4
X=pd.DataFrame([data['Salary'],data['Salary']**2,data['Salary']**3,data['Salary']**4]).T
y=data['Level']
model.fit(X,y)
# bac 4
X = pd.DataFrame([data['Salary'],data['Salary']**2,data['Salary']**3,data['Salary']**4]).T # T chuyen vi )
print('Intercept cua mo hinh :',model.intercept_)
print('Coefficient cua mo hinh :',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,y=data['Level']))


import numpy as np
data.plot(kind='scatter',x='Salary',y='Level',figsize=(9,9),color = 'red',xlim=(0.7))
range =np.arange(data['Salary'].min(),data['Salary'].max(),10) # coi lai cho ve do thi 
predictor = pd.DataFrame([range,range**2,range**3,range**4]).T
y_pred = model.predict(X=predictor)
plt.plot(range,y_pred,color='blue')
```
